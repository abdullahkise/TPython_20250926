### urlopen

```python
import requests
```

### get

```python
rget = requests.get("https://www.python.org/jobs/")

dir(rget)
#
rget.status_code #200 kodu başarılı demektir.
rget.elapsed.total_seconds() #0.353755

#
webSayfasi = rget.content.decode() #web sayfaları bytestr şeklinde gelir.okunması için decode edelim.

print(webSayfasi[:100]) #ilk 100 karakter gelsin.
```

## BeatifulSoup Kütüphanesi
![alt text](BeatifulSoup.png)

```python
from bs4 import BeautifulSoup
```

```python
html_metin = requests.get("https://www.python.org/jobs/").content.decode()
len(html_metin)
```

```python
soup = BeautifulSoup(html_metin, "html.parser")

type(soup)
#dir(soup)
```

```python
#ilk etiketi getirir.
soup.title
soup.body.h1
soup.body.h2

#bu hiyerarşideki tüm arananlar.
soup.body("h2")

#
soup.body.section.h2.a.string #ilk görüğü metni getirr.
soup.body.section.h2.text #alt hiyerarşideki textleri getirir
```

```python
for link in soup.body.section("a")[:3]:
    #print(type(link)) #<class 'bs4.element.Tag'>
    #print(link.text)
    #print(link["href"])
    #print(link.get("href"))

    print("{:<30.30s} {:>70}".format(link.text, link["href"]))
```

### find ve find_all
![alt text](Bs4BirdenFazlaEtiket.png)


```python
soup.find("a") #ilk bulduğu a etiketi
soup.find_all("a") #tüm a etiketleri
soup.find_all("a",attrs={"class":"jump-to-menu"},limit=1)
```
